```python
# Importar as bibliotecas

import json
import os
import pandas
import urllib.request
```

```python
# Traz as opções de configuração do JSON

json_file = os.path.abspath('../../options.json')

with open(json_file, 'r', encoding='utf-8') as j_file:
    json_data = json.load(j_file)
```

```python
# Obtenção da planilha com os dados diários do CDI

cvm_path = os.path.join(os.path.abspath('../../archives/'), 'cdi-data', 'cdi_data.csv')
df_cdi = pandas.read_csv(cvm_path, sep=';')
```

```python
# Filtragem pela data dos dados

START_DATE = json_data['CONFIG']['START_DATE']
END_DATE = json_data['CONFIG']['END_DATE']

df_cdi = df_cdi.query("@START_DATE <= DATE <= @END_DATE").copy()
```

```python
# Valorização do cdi

list_cdi = (df_cdi['CDI_TAX']).to_list()
list_cdi_valuation = []
last_cdi_tax = 0

for cdi_tax in list_cdi_valuation:
    if last_cdi_tax != 0:
        cdi_tax = ((cdi_tax / 100) + 1) * last_cdi_tax
    else:
        cdi_tax = ((cdi_tax / 100) + 1)
    list_cdi_valuation.append(cdi_tax)
    last_cdi_tax = cdi_tax

print(last_cdi_tax)
```
